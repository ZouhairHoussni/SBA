#### Création d'un dictionnaire pour imputer des valeurs manquantes : ( city - state )

```python
import pandas as pd

df = pd.read_csv("uszips.csv")

# choisir les colonnes à garder
columns_to_keep = ['zip', 'city', 'state_id','state_name']

df_zip = df[columns_to_keep]

# transformer la colonne "zip" en chaine de caractères 
# zfill() ajoutte des 0 à la chaine de caractère
df_zip["zip"] = df_zip['zip'].astype(str).str.zfill(5)

# création d'un dictionnaire qui a comme clé "zip" et valeurs
city_dict = df_zip.set_index('zip')['city'].to_dict()


state_dict = df_zip.set_index('zip')['state_name'].to_dict()


```
