```python
df_zip = pd.read_csv("uszips.csv")

# choisir les colonnes à garder
columns_to_keep = ['zip', 'city', 'state_id','state_name']

df_zip = df_zip[columns_to_keep]

# transformer la colonne "zip" en chaine de caractères 
# zfill() ajoutte des 0 à la chaine de caractère
# df_zip["zip"] = df_zip['zip'].astype(str).str.zfill(5)

# création d'un dictionnaire qui a comme clé "zip" et valeur "city"
city_dict = df_zip.set_index('zip')['city'].to_dict()

# création d'un dictionnaire qui a comme clé "zip" et valeur "State_id"
state_dict = df_zip.set_index('zip')['state_id'].to_dict()

```

```python
df[df["City"].isna()]
```

```python
import numpy as np

df['City'] = df.apply(lambda x: 
                            city_dict.get(x['Zip'], np.nan)
                            if pd.isna(x['City']) else x['City'], axis=1)
```

```python
df['State'] = df.apply(lambda x: 
                            state_dict.get(x['Zip'], np.nan)
                            if pd.isna(x['State']) else x['State'], axis=1)
```

```python
df.isna().sum()
```

```python
df1 = df[df["DisbursementDate"].isna()]

df1
```

```python
mask1 = df["Bank"].notna()

mask2 = df["BankState"].isna()

df.loc[mask1 & mask2,"BankState"] = "Out"

df[df["BankState"]=="Out"]
```

```python
df.isna().sum()
```

```python
df["RevLineCr"].value_counts()
```

```python

df["RevLineCr"] = df["RevLineCr"].replace({"T":"Y" , "0": "N"})

df = df[df['RevLineCr'].isin(["N", "Y"])]

df["RevLineCr"].value_counts()
```

```python
df.isna().sum()
```

```python
df = df.dropna(subset=['MIS_Status','Bank','State','City','Name'])

df.isna().sum()
```

```python

df["NewExist"] = df["NewExist"].replace({ 0 : 1, "Y":1, "N" : 2})

df["NewExist"].value_counts()
```

```python
df["LowDoc"].value_counts()
```

```python
df = df[df['LowDoc'].isin(["N", "Y"])]
```

```python
df = df.dropna(subset=['LowDoc','NewExist'])

df.isna().sum()
```

```python
df["FranchiseCode"].value_counts()
```

```python

df["FranchiseCode"] = df["FranchiseCode"].replace({0:"N",1:"N"})

mask_Franchise = df["FranchiseCode"] != "N"

df.loc[mask_Franchise,"FranchiseCode"] = "Y"

df["FranchiseCode"].value_counts()
```

```python
df = df.drop(columns=["LoanNr_ChkDgt","Name","Zip","ApprovalDate","ApprovalFY","ChgOffDate","DisbursementDate","DisbursementGross","BalanceGross","ChgOffPrinGr"])

df
```

```python
df.info()
```

```python
df['SBA_Appv'] = df['SBA_Appv'].str.replace("\$",'')

df['GrAppv'] = df['GrAppv'].str.replace("\$",'')

df['SBA_Appv'] = df['SBA_Appv'].str.replace("\,",'')

df['GrAppv'] = df['GrAppv'].str.replace("\,",'')

df['NewExist'] = pd.to_numeric(df["NewExist"], errors='coerce',downcast='integer')
```

```python
df.head()
```

```python
df = df.astype(object)
df["Term"] = pd.to_numeric(df["Term"], errors='coerce',downcast='integer')
df["NoEmp"] = pd.to_numeric(df["NoEmp"], errors='coerce',downcast='integer')
df["GrAppv"] = pd.to_numeric(df["GrAppv"], errors='coerce',downcast='integer')
df["SBA_Appv"] = pd.to_numeric(df["SBA_Appv"], errors='coerce',downcast='integer')
df["CreateJob"] = pd.to_numeric(df["Term"], errors='coerce',downcast='integer')
df["RetainedJob"] = pd.to_numeric(df["Term"], errors='coerce',downcast='integer')

```

```python
df.to_csv("SBA_Cleaned.csv", index=False)
```
